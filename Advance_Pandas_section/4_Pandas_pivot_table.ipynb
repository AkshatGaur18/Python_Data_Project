```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# Counts of the job Titles

```python
df.groupby('job_title_short')['job_title_short'].count()
```

```python
df.groupby('job_title_short')['job_title_short'].size()
```

```python
df.pivot_table(index='job_title_short',aggfunc='size')
```

```python
df.pivot_table(index='job_title_short',values='salary_year_avg',columns='job_country',aggfunc='median')
```

```python
df.pivot_table(index='job_country',values='salary_year_avg',columns='job_title_short',aggfunc='median')
```

```python
df['job_country'].value_counts()
```

```python
df['job_country'].value_counts().head(6)
```

```python
top_countries = df['job_country'].value_counts().head(10).index
```

```python
df_job_country_salary = df.pivot_table(index='job_country',columns='job_title_short',values='salary_year_avg',aggfunc='median')
```

```python
df_job_country_salary
```

```python
job_title = ['Data Analyst','Data Engineer','Data Scientist']
```

```python
df_job_country_salary[job_title]
```

```python
df_job_country_salary = df_job_country_salary.loc[top_countries,job_title]
```

```python
df_job_country_salary.plot(kind='bar')
plt.ylabel('Median Salary')
plt.xlabel('Top Countries')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45,ha = 'right')
plt.show()
```
