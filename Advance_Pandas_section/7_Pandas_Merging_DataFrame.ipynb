```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df[df['job_country'] == 'United States']
```

```python
df_usa = df[df.job_country == 'United States'].copy()
```

```python
df_usa['job_posted_date'].dt.strftime('%B')
```

```python
df_usa['job_posted_month'] = df_usa['job_posted_date'].dt.strftime('%B')
df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'],format = '%B').dt.month
df_usa_pivot
```

```python
df_usa_pivot.sort_values('month_no',inplace=True)
df_usa_pivot.set_index('job_posted_month',inplace=True)
df_usa_pivot.drop(columns='month_no',inplace=True)
df_usa_pivot
```

```python
df_usa_pivot.plot(kind='line')
```

```python
df_usa
```

```python
top = df_usa['job_title_short'].value_counts().head(4)
top.index
```

```python

top = top.index.to_list()
top
```

```python
df_usa_pivot[top].plot(kind='line')
```

```python
df_usa_pivot
```

```python
df_usa_software = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_usa_software
```

```python
print(df_usa_pivot.merge(df_usa_software,on='job_posted_month'))
df_us_merge = df_usa_pivot.merge(df_usa_software,on='job_posted_month')
```

```python
top_5 = (df_us_merge
         .sum()
         .sort_values(ascending=False)
         .head()
         .index
         .to_list())
top_5
```

```python
df_us_merge[top_5].plot(kind='line')
plt.title('Monthly jobs Posting for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Counts')
plt.legend()
plt.ylim(0,22000)
plt.show()
```
