```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

# Filter for INDIA Data Analyst roles


```python
df_DA_US = df[(df['job_country']== 'United States') & (df['job_title_short']=='Data Analyst')]
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).to_frame()

df_plot
```

```python
import seaborn as sns
df_plot = df_plot.sort_values(by='count',ascending=False)
sns.barplot(data=df_plot,x='count',y='job_location',hue='count',palette='dark:b_r')
sns.despine()
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.title('Counts of job Location for Data Analyst in the US')
plt.legend().remove()

plt.show()
```

```python
dict_column = {
    'job_work_from_home':'Work from Home Offered',
    'job_no_degree_mention':'Degree Requirement',
    'job_health_insurance':'Health Insurance Offered'
}

fig , ax = plt.subplots(1,3)
fig.set_size_inches((12,5))

for i,(columns,title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[columns].value_counts(),labels = ['False','True'] ,startangle=90, autopct='%1.1f%%')
    ax[i].set_title(title)

plt.show()
```

```python
df_company = df_DA_US['company_name'].value_counts().head(10).to_frame()
df_company = df_company.sort_values(by='count',ascending=False)
sns.barplot(data=df_company,x='count',y='company_name',hue='count',palette='dark:b_r')
sns.despine()
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.title('Counts of job Location for Data Analyst in the US')
plt.legend().remove()

plt.show()
```
