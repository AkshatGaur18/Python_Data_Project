```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title'] =='Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month'] = df_DA_US['job_posted_date'].dt.month
```

```python
df_DA_US_Explode = df_DA_US.explode('job_skills')
```

```python
df_DA_US_Pivot = df_DA_US_Explode.pivot_table(index='job_posted_month' , columns='job_skills' , aggfunc='size' ,fill_value=0)
```

```python
df_DA_US_Pivot.loc['Total'] = df_DA_US_Pivot.sum()
df_DA_US_Pivot
```

```python
df_DA_US_Pivot = df_DA_US_Pivot[df_DA_US_Pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_US_Pivot = df_DA_US_Pivot.drop('Total')
```

```python
df_DA_US_Pivot
```

```python
DA_Totals = df_DA_US.groupby('job_posted_month').size()
```

```python
df_DA_US_percentage = df_DA_US_Pivot.div(DA_Totals/100,axis=0)
df_DA_US_percentage
```

```python
df_DA_US_percentage = df_DA_US_percentage.reset_index()
df_DA_US_percentage['job_posted_month_name'] = df_DA_US_percentage['job_posted_month'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percentage = df_DA_US_percentage.set_index('job_posted_month_name')
df_DA_US_percentage = df_DA_US_percentage.drop(columns= 'job_posted_month')
```

```python
df_plot = df_DA_US_percentage.iloc[:, :5]

sns.lineplot(data=df_plot,dashes=False,palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending top Skills for Data Analysts in the US')
plt.ylabel('Likelyihood in the job posting')
plt.xlabel('2023')
plt.legend().remove()
from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))
for i in range(5):
    plt.text(11.2,df_plot.iloc[-1,i],df_plot.columns[i])
```
