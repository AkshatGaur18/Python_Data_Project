```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie',startangle=90 ,autopct='%1.1f%%')
plt.title('People Who Are doing Work From Home')
plt.ylabel('')
```

```python
df[['job_work_from_home','job_no_degree_mention','job_health_insurance']]
```

```python
job = {'job_work_from_home':'Work From Home',
       'job_no_degree_mention':"Job Degree Req",
       'job_health_insurance':'Health Insurance Offered'}
fig ,ax = plt.subplots(1,3)
for i,(column,title) in enumerate(job.items()):
    df[column].value_counts().plot(kind='pie',startangle=90,autopct='%1.1f%%',ax=ax[i],title=title,ylabel='')
    fig.tight_layout()
```
