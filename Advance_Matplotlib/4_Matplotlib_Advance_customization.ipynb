```python
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x:ast.literal_eval(x) if pd.notna(x) else x)

df_DA = df[df['job_title_short']=='Data Analyst'].copy()
```

```python
df_DA
```

```python
df_DA['job_posted_month'] = df['job_posted_date'].dt.month
df_DA
```

```python
df_DA_explode = df_DA.explode('job_skills')
```

```python
df_DA_explode
```

```python
df_DA_pivot = df_DA_explode.pivot_table(index='job_posted_month' ,columns='job_skills' ,aggfunc='size',fill_value=0)
df_DA_pivot
```

```python
df_DA_pivot.loc['Total'] = df_DA_pivot.sum()
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc['Total'].sort_values(ascending=False).index]
df_DA_pivot = df_DA_pivot.drop('Total')
```

```python
df_DA_pivot = df_DA_pivot.reset_index()
df_DA_pivot
```

```python
df_DA_pivot['job_posted_month'] = df_DA_pivot['job_posted_month'].apply(lambda x  : pd.to_datetime(x,format='%m').strftime('%B'))
df_DA_pivot
```

```python
df_DA_pivot = df_DA_pivot.set_index('job_posted_month')
df_DA_pivot
```

```python
df_DA_pivot.iloc[:,:7].plot(
    kind='line',
    linewidth = 4,
    linestyle = '-',#'-','--','-.',':'
    colormap= 'viridis',  #'viridis' #'plasma', 'inferno' , 'magma' , 'cividis',
    marker = 'o',
    markersize=7,
    figsize=(15,10)
    
    
    )
plt.title('Top 5 Skills for Data Analyst Per Month')
plt.ylabel('Count')
plt.xlabel('')
plt.show()
```

```python

# Filter data for 'Data Analyst' jobs
df = df[df['job_title_short'] == 'Data Analyst']

# Explode the job skills column
df_exploded2 = df.explode('job_skills')

# Group by job_skills to get skill count and median salary
skills_stats = df_exploded2.groupby('job_skills').agg(
    skill_count=('job_skills', 'count'),
    median_salary=('salary_year_avg', 'median')
)

# Sort and select top 20 skills
skills_stats = skills_stats.sort_values(by='skill_count', ascending=False).head(20)

# Create scatter plot
fig, ax = plt.subplots()
skills_stats.plot(kind='scatter', x='skill_count', y='median_salary', ax=ax)

# Add title and labels
plt.title('Salary vs. Count of Job Postings for Top 20 Skills')
plt.xlabel('Count of Job Postings')
plt.ylabel('Median Yearly Salary')

# Annotate points with skill names
texts = []
for i, txt in enumerate(skills_stats.index):
    texts.append(plt.text(skills_stats['skill_count'].iloc[i], skills_stats['median_salary'].iloc[i], txt))

# Adjust text to avoid overlap
adjust_text(texts, arrowprops=dict(arrowstyle="-", color='b', lw=1))

# Format y-axis (salary) labels
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, pos: f'${int(y/1000)}k'))

# Tight layout for better spacing
plt.tight_layout()

# Show plot
plt.show()
```
